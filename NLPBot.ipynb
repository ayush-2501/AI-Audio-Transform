```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline
import spacy
from abc import ABC, abstractmethod

# Load NLP Models
nlp = spacy.load("en_core_web_trf")
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
zero_shot_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Skills and their embeddings
skills = ["leave"]  # More skills can be added later (e.g., "payroll")
skill_embeddings = embedding_model.encode(skills, convert_to_tensor=False)
dimension = skill_embeddings.shape[1]

# FAISS index for fast skill retrieval
faiss_index = faiss.IndexFlatL2(dimension)
faiss_index.add(np.array(skill_embeddings, dtype=np.float32))


# ====== Skill Base Class ======
class Skill(ABC):
    def __init__(self, name, intents):
        self.name = name
        self.intents = intents

    @abstractmethod
    def extract_entities(self, query):
        pass

    def classify_intent(self, query):
        """Classify intent using zero-shot learning within the skill-specific intents."""
        intent_result = zero_shot_classifier(query, candidate_labels=list(self.intents.keys()))
        return list(self.intents.keys())[intent_result["labels"].index(intent_result["labels"][0])]


# ====== LEAVE SKILL ======
class LeaveSkill(Skill):
    leave_types = [
        "Annual Leave", "Maternity Leave", "Personal Leave",
        "Sick Leave", "Medical Leave", "Emergency Leave"
    ]

    reasons = {
        "Sick": ["Unwell", "Sick"],
        "Back Pain": ["Back Pain"],
        "Cold": ["Cold"],
        "Eye Infection": ["Eye Infection"],
        "Flu": ["Flu"],
        "Food Poisoning": ["Food Poisoning"],
        "Illness": ["Fever", "Headache", "Head Ache"]
    }

    reason_embeddings = embedding_model.encode(list(reasons.keys()), convert_to_tensor=False)

    required_entities = {
        "apply_leave": ["leave_type", "reason", "start_date", "end_date"],
        "cancel_leave": ["start_date", "end_date"],
        "leave_balance": []
    }

    def __init__(self):
        intents = {
            "apply_leave": "Requesting or applying for leave",
            "cancel_leave": "Canceling or withdrawing leave request",
            "leave_balance": "Checking remaining leave balance"
        }
        super().__init__("leave", intents)

    def extract_entities(self, query):
        """Extract leave-specific entities."""
        doc = nlp(query)
        entities = {"reason": "", "leave_type": "", "start_date": "", "end_date": ""}

        for ent in doc.ents:
            if ent.label_ == "DATE":
                if not entities["start_date"]:
                    entities["start_date"] = ent.text
                elif not entities["end_date"]:
                    entities["end_date"] = ent.text

        entities["reason"] = self.extract_reason(query)

        leave_type_result = zero_shot_classifier(query, candidate_labels=self.leave_types)
        entities["leave_type"] = leave_type_result["labels"][0] if leave_type_result["scores"][0] > 0.7 else ""

        return entities

    def extract_reason(self, query):
        """Extract reason using predefined keywords or embedding similarity."""
        for reason, variations in self.reasons.items():
            for var in variations:
                if var.lower() in query.lower():
                    return reason

        query_embedding = embedding_model.encode([query], convert_to_tensor=False)
        scores = np.dot(self.reason_embeddings, query_embedding.T).flatten()
        best_match_index = np.argmax(scores)

        return list(self.reasons.keys())[best_match_index] if scores[best_match_index] > 0.7 else ""

    def check_missing_entities(self, intent, extracted_entities):
        """Check for missing required entities and return a list of missing ones."""
        missing_entities = [entity for entity in self.required_entities[intent] if not extracted_entities.get(entity)]
        return missing_entities


# Skill Registry
skill_registry = {
    "leave": LeaveSkill()
}


def extract_skills_and_intent(query):
    """Extract skill, intent, and entities based on detected skill."""
    query_embedding = embedding_model.encode([query], convert_to_tensor=False)
    _, indices = faiss_index.search(np.array(query_embedding, dtype=np.float32), k=1)
    extracted_skill = skills[indices[0][0]] if indices[0][0] < len(skills) else "unknown"

    if extracted_skill in skill_registry:
        skill_obj = skill_registry[extracted_skill]
        extracted_intent = skill_obj.classify_intent(query)
        entities = skill_obj.extract_entities(query)
        missing_entities = skill_obj.check_missing_entities(extracted_intent, entities)

        return extracted_skill, extracted_intent, entities, missing_entities
    else:
        return "unknown", "unknown", {}, []


# === Chatbot Flow ===
def chatbot():
    print("Welcome! How can I assist you today?")

    for data in sample_data:
        session_entities = {}  # Reset entities for each query
        user_input = input("You: ")

        skill, intent, entities, missing_entities = extract_skills_and_intent(user_input)
        session_entities.update(entities)

        while missing_entities:
            next_missing = missing_entities[0]
            user_response = input(f"I need more information. What is your {next_missing}? : ")  # Ask user
            session_entities[next_missing] = user_response

            # Check again for any remaining missing entities
            missing_entities = [key for key, value in session_entities.items() if not value]

        print("Processing your request...")
        print(f"Skill: {skill}")
        print(f"Intent: {intent}")
        print(f"Entities: {session_entities}")
        print("Request submitted successfully!\n")

if __name__ == "__main__":
    chatbot()
```
